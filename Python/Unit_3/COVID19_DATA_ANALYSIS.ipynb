# 2. Understand COVID 19 in the 5 states since December 22nd

## a. Calculate new cases(different approach):

Define a function for calculating new cases:

number of new cases of today = total number of cases by the end of today - total number of cases yesterday by the end of yesterday  

```
def get_new_cases(df):
  new_cases = df['cases']-df['cases'].shift(1)
  new_cases.iloc[0] = df['cases'].iloc[0]
  return new_cases
```

Calculate new cases for the 5 states using a for loop and the function defined previously 

```
top_5_states=[]
for name in names:
  state_df = df[df['state']==name]
  state_df['new_cases']=get_new_cases(state_df)
  top_5_states.append(state_df)
top_5_states = pd.concat(top_5_states)
```

```
top_5_states.head(5)
```

```
top_5_states.tail(5)
```

Visualize number of new cases in the top 5 states

```
plt.figure(figsize=(15,7))
for state in names:
  df_st = top_5_states[(top_5_states['state']==state) & (top_5_states['date']>='2020-12-22')]
  plt.plot(df_st['date'],
           df_st['new_cases']
           )
  plt.xticks(rotation=90)
  plt.title("Number of new cases in the top 5 states")
plt.legend(names)
plt.show()
```

From the graph above, we see that Calfornia had highest number of new cases almost all the days we investigated (on January 4th, New Jersey had the highest number of new cases). 

For New York, we see that the curve is quite flat but with noticeable increment, from about 10000 new cases a day to 20000 new cases a day. 

## b. Calculate new deaths (different approach):

Define a function similar to calculate new cases. 



```
def get_new_deaths(df):
  new_deaths = df['deaths']-df['deaths'].shift(1) # today's death - yesterday's death(obtain by shifting the deaths column down by one
  new_deaths.iloc[0] = df['deaths'].iloc[0]
  return new_deaths
```

Calculate new deaths for the 5 states

```
top_5_states_with_new_deaths=[]
for name in names:
  state_df = top_5_states[top_5_states['state']==name]
  state_df['new_deaths']=get_new_deaths(state_df)
  top_5_states_with_new_deaths.append(state_df)
top_5_states_with_new_deaths = pd.concat(top_5_states_with_new_deaths)
```

```
top_5_states_with_new_deaths.head(5)
```

```
top_5_states_with_new_deaths.tail(5)
```

```
plt.figure(figsize=(15,7))
for state in names:
  df_st = top_5_states_with_new_deaths[(top_5_states_with_new_deaths['state']==state) & (top_5_states_with_new_deaths['date']>='2020-12-22')]
  plt.plot(df_st['date'],
           df_st['new_deaths']
           )
  plt.xticks(rotation=90)
  plt.title("Number of new deaths in the top 5 states")
plt.legend(names)
plt.show()
```

From the graph, we see that California again has the highest number of deaths for most of the days we investegated. The maximum number of new deaths was 674 on 2021-01-08

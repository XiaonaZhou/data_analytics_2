## Slide 18 

Write an if-elif-else chain to determine the persons age group.

- For example, if the person is less than 2 years old, print that person is a baby.
- This should work on someone that is over 70!




```
age = int(input("Enter your age: "))
if age<2:
  print("A baby")
elif age<12:
  print("A kid")
elif age < 21: 
  print("A teen")
else:
  print("An adult") 
```

Write a grocery list.  Then write a series of if statements that check if the items in your cart are on your list. 

- You should have at least 5 if statements. 

```
grocery_list = ["apple", "rice", "beef", "grape","mango" ]
car_list = ["beef", "grape", "cookies", "mango", "pencil"]
```

```
for items in car_list:
  if items in grocery_list:
    print("You have "+ items)
  else:
    print("You shouldn't buy " + items + ", put it back! ")
```
